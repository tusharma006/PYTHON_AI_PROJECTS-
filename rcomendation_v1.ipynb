```python
from langchain_core.documents import Document
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings

```

```python
import json
with open("songs.json", "r", encoding="utf-8") as f:
    songs = json.load(f)
print(f"Total Songs: {len(songs)}")
print(songs)
```

```python
document = []

for song in songs:

    page_content = f"""
Song Name: {song['song_name']}
Artist: {song['artist']}
Genre: {song['genre']}
Mood: {song['mood']}
Tempo: {song['tempo']}
Language: {song['language']}
Release Year: {song['release_year']}
Keywords: {", ".join(song['keywords'])}
Description: {song['description']}
"""
    document.append(
        Document(
            page_content=page_content,#page_content=str(song)but this is not ideal  beacause Embedding models are trained primarily on natural language sentences, not Python dictionary syntax.
            metadata=song
        )
    )

# print("Total Documents :", len(document))
# for i in document:
#   print(i.metadata)
# # print(document[].metadata)
```

```python
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
vector_db = Chroma.from_documents(
    documents=document,# = songs is list of dic and choma expect document eit page_content and meta data
    embedding=embedding_model,
    persist_directory="chroma_db"# When Chroma creates embeddings, it stores them somewhere. i we haven't this it stays in ram and when colab resatrts everything is  lost
)
print(len(document))
print("Total Documents :", vector_db._collection.count())
```

```python
print("\nAvailable Songs\n")
for i, song in enumerate(songs, start=1):
    print(f"{i:02d}. {song['song_name']}  -  {song['artist']}")
```

```python
query = input("""enter the query:""")

# results = vector_db.similarity_search_with_relevance_scores(
#     query=query,
#     k=5
# )

results = vector_db.similarity_search_with_score(
    query=query,
    k=5
)
results
print(f"Total Results: {len(results)}")
```

```python
# page_content is for emmbeding and meta data is to access particular field after search
for i, (doc,score) in enumerate(results,start=1,):

    print("----"*40)
    print(f"Recommendation {i}")
    print("Song:", doc.metadata["song_name"])
    print("Artist:", doc.metadata["artist"])
    print("Genre:", doc.metadata["genre"])
    print("Mood:", doc.metadata["mood"])
    print("Language:", doc.metadata["language"])
    print("Release Year:", doc.metadata["release_year"])
    print("Description:", doc.metadata["description"])
    print(f"score:{score}")
     #  print(i)
  #  print(doc) we should use clean approch, this will print entire object
  #  print(type(doc))

```
